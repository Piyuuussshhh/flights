So a value of 628 indicates that there are 322 + x = 628 unique numbers that are used in place of airport iata codes in the month of October; x = 306.
- We can map the 302 numbers to 302 unique IATA codes to generate factually false but type-wise correct data.

```python
from constants import IATA_CODE

airports = pd.read_csv("../datasets/airports.csv")
iata_codes = airports[IATA_CODE].tolist()

def map() -> dict:
    seen = set()
    selected_codes = []
    for iata_code in iata_codes:
        if iata_code not in seen:
            seen.add(iata_code)
            selected_codes.append(iata_code)
        if len(selected_codes) == len(invalid_nums):
            break
    return dict(zip(invalid_nums, selected_codes))

invalid_nums_mapping = map()

flights[ORIGIN_AIRPORT] = flights[ORIGIN_AIRPORT].replace(invalid_nums_mapping)
flights[DESTINATION_AIRPORT] = flights[DESTINATION_AIRPORT].replace(invalid_nums_mapping)
```
